```python
import pandas as pd

data = pd.read_csv('/Users/baby_cow/Desktop/AI/ML_competition/data/geocoded_addresses_5.csv')
```

# 지역을 나누자

```python
data['구'].unique()
```

```python
from IPython.display import display

region_chunks = {}
for region in data['구'].unique():
    temp = data[data['구'] == region]
    region_chunks[region] = temp

display(region_chunks['강남구'])
```

#  시간순으로 정렬

```python
for region in data['구'].unique():
    chunk = region_chunks[region]
    chunk = chunk.sort_values(by='계약년월')
    region_chunks[region] = chunk
```

# 동은 년도별로 나눈 후 target의 mean 값을 사용하자

```python
chunk = region_chunks['강남구']

chunk
```

```python
chunk['계약년'] = chunk['계약년월'].map(lambda x : int(str(x)[:4]))
year_chunks = {}
for year in chunk["계약년"].unique():
    year_chunks[year] = chunk[chunk['계약년'] == year]

chunk['계약년'].unique()
```

```python
year_chunks[2023][['target', "계약년월", '계약년', '동', '구']]
```

# 동 -> 그해 그 동의 target 평균가

```python
for year in range(2007, 2024):
    year_data = year_chunks[year]

    for region in year_data['동'].unique():
        train_region_df = year_data[year_data['동'] ==  region].query('is_test == 0')
        dong_value = train_region_df['target'].mean()
        year_data.loc[year_data['동'] == region, '동'] = dong_value

    year_chunks[year] = year_data

year_chunks[2023]

```
